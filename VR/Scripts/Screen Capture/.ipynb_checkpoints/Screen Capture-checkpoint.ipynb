```python
import pyautogui
import win32gui

def screenshot(window_title=None):
    if window_title:
        hwnd = win32gui.FindWindow(None, window_title)
        if hwnd:
#             win32gui.SetForegroundWindow(hwnd)
            x, y, x1, y1 = win32gui.GetClientRect(hwnd)
#             print(x, y, x1, y1)
            x, y = win32gui.ClientToScreen(hwnd, (x, y))
            x1, y1 = win32gui.ClientToScreen(hwnd, (x1 - x, y1 - y))
            im = pyautogui.screenshot(region=(x, y, x1, y1))
            return im
        else:
            print('Window not found!')
    else:
        im = pyautogui.screenshot()
        return im


# display(im)
# if im:
#     im.show()
```

```python
import requests
import ast
import base64
import json
def detect_object(img_path, obj_type):
    url = 'https://hackaroo.ngrok.io/api/predict/house'
    
    if obj_type == "road":
        url = 'https://hackaroo.ngrok.io/api/predict/road'
    elif obj_type == "vehicle":
        url = 'https://hackaroo.ngrok.io/api/predict/vehicle'
    elif obj_type == "utility":
        url = 'https://hackaroo.ngrok.io/api/predict/utility'
    files = {'file': open(img_path, 'rb')}
    response = requests.post(url, files=files)
    predicted_img = json.loads(response.text)
    print(predicted_img)
    if (predicted_img['image'] is not None):
        imgdata = base64.b64decode(predicted_img['image'])
        filename = 'predicted.jpg'  # I assume you have a way of picking unique filenames
        with open(filename, 'wb') as f:
            f.write(imgdata)
            return True
    return False

        
```

```python
import tkinter as tk
from PIL import ImageTk, Image

#This creates the main window of an application
window = tk.Tk()
window.title("Google Earth VR (with Object Detection)")
window.geometry("640x480")
window.configure(background='grey')

path = "test.jpg"
earth_screenshot = None
# Creates a Tkinter-compatible photo image, which can be used everywhere Tkinter expects an image object.
# img = ImageTk.PhotoImage(earth_screenshot)
img_2 = ImageTk.PhotoImage(Image.open(path))
# img_3 = ImageTk.PhotoImage(Image.open("test2.jpg"))

tkimg = [None]
#The Label widget is a standard Tkinter widget used to display a text or image on the screen.
panel = tk.Label(image = img_2)
panel.pack(side = "bottom", fill = "both", expand = "yes")
def change_image():
    '''
    Get new screenshot
    '''
    earth_screenshot = screenshot('Google Earth VR')
    earth_screenshot.save("test_VR.jpg")
    
#     earth_im = ImageTk.PhotoImage(earth_screenshot)
    '''
    Predict screenshot via API
    '''
    detect_object('test_VR.jpg', 'house')
    
    predicted_img = ImageTk.PhotoImage(Image.open('predicted.jpg'))
    tkimg[0] = predicted_img       
    
    panel.configure(image=tkimg[0])
    window.update_idletasks()
    window.after(2000, change_image)
change_image()     
# window.bind("<Return>", callback)
#Start the GUI
window.mainloop()
```
