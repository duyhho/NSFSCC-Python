## Group By Call Category



```python
from tqdm.notebook import tqdm
CATEGORY_MAP = {
    'City Facilities': ['City Facilities', 'City Clerks Office','City Managers Office','City Planning & Development', 'Facilities'],
    'Parking': ['Park','Park Maintenance','Parking Meter','Downtown Parking'],
    'Property Violations': ['Property / Buildings / Construction', 'Property Preservation','Property Violations', 'Graffiti', 'Property & Nuisance Violations', 'Nuisance Violations'],
    'Animals': ['Animals / Pets', 'Animal'],
    'Water': ['Water','Water Leak','Storm Water','Storm Water / Sewer','Wastewater','Water Main Break','Water Quality','Water Services','Sewer','Ditch'],
    'Trash': ['Trash / Recycling','Trash'],
    'Public Health': ['Health','Public Health','Food Protection','Air Quality','Weeds'],
    'Lights/Signals/Signs': ['Lights / Signals','Signal','Signs'],
    'Vehicles': ['Vehicle', 'Vehicles / Parking'],
    'Street/Sidewalks': ['Street', 'Street Light','Street Sign','Pothole','Sidewalk','Sidewalks / Curbs / Ditch','Streets / Roadways / Alleys','Steel Plate','Markings'],
    'Legal': ['Government','Mayors Office','Permit','Law','Municipal Court'],
    'Public Safety': ['Public Safety','Fire','Snow & Ice','Trees'],
    'Parks and Recreation': ['Parks & Recreation','Boulevard Parks and Rec'],
    'Neighborhood': ['Neighborhood & Community Services', 'Neighborhood Preservation'],
    'Data Not Available': ['Data Not Available'],
    'Information Request': ['Information Request'],
    'Maintenance': ['Maintenance','Bridge','Guard Rail','Cleaning'],
    'Traffic': ['Traffic Sign','Traffic Study'],
    'Information Technology': ['Information Technology'],
    'Noise': ['Noise Control'],
    'Other': ['Aviation','Public Works','Zoning','Right of Way','Sealing','Sweeping','Bulky Appointment',
              'Consumer Services','Convention & Entertainment Center','Finance','Housing','Mowing / Weeds','Capital Projects']
}
NEW_CATEGORIES = []
for index, row in tqdm(sample_df.iterrows(), total = sample_df.shape[0]):
  new_cat = None
  category = row['CATEGORY']
  for key, value in CATEGORY_MAP.items():
    if category in value:
      new_cat = key
      NEW_CATEGORIES.append(new_cat)
  if new_cat is None:
    print(category)
  assert new_cat is not None
  # if index == 10:
  #   break
sample_df['CATEGORY_NEW'] = NEW_CATEGORIES
  
```

```python
'''
Group By Call Category
'''
import numpy as np
from sklearn.preprocessing import StandardScaler
call_df = sample_df.groupby(['CATEGORY_NEW', 'nbh_id']).size()
call_df = call_df.unstack(level = 0).fillna(0).astype(int) #level = 1 will swap column and row
# call_df['Total'] = call_df.sum(axis=1)#; call_df.loc['Total'] = call_df.sum()
# call_df[call_df.columns] = call_df[call_df.columns].applymap(int)

call_df_org = call_df.copy()
call_df = call_df.loc[:,"Animals":"Water"].div(call_df.sum(axis=1), axis=0)
call_df = call_df.merge(socio_df, how='inner', left_on='nbh_id', right_on='NBH_ID')#.drop(columns=['Neighborhood Number'])
call_df.index = call_df['NBH_ID']
# call_df.loc[:,"Animals":"Water"] = temp_df*100


call_df.head()
```

## Group By Response Time



```python
import plotly.express as px
import numpy as np
from tqdm.notebook import tqdm
RESPONSE_THRESHOLD = {
    0: 'On time (0-day delay)',
    1: '1 day',
    2: '2 days',
    7: '7-day window',
    14: '14-day window',
    30: '30-day window',
    60: '60-day window',
    90: '90-day window',
    120: '120-day window',
    180: '180-day window',
    365: '365-day window', 
    4525: '365+ -day window'
    }

# response_count_df = sample_df['DAYS TO CLOSE'].to_frame('DAYS') 
# response_count_df.head()
# counts, bins = np.histogram(response_count_df['DAYS'], bins=list(RESPONSE_THRESHOLD.keys()))
# bins = 0.5 * (bins[:-1] + bins[1:])
# fig = px.bar(x=bins, y=counts, labels={'x':'DAYS', 'y':'count'})
# fig.show()

new_days_to_close = []
for index, row in tqdm(sample_df.iterrows(), total = sample_df.shape[0]):
  old_days = row['DAYS TO CLOSE']
  temp = len(new_days_to_close)
  # print(old_days)
  for lim in RESPONSE_THRESHOLD:
    if old_days <= lim:
      # print(RESPONSE_THRESHOLD[lim])
      new_days_to_close.append(RESPONSE_THRESHOLD[lim])
      break
  if temp == len(new_days_to_close):
    display(row)
  # break
sample_df['DAYS TO CLOSE (NEW)'] = new_days_to_close
sample_df[['DAYS TO CLOSE','DAYS TO CLOSE (NEW)']].head()
```

```python
'''
Group By Response Time Category
'''
response_df = sample_df.groupby(['DAYS TO CLOSE (NEW)', 'nbh_id']).size()
response_df= response_df.unstack(level = 0).fillna(0).astype(int) #level = 1 will swap column and row
response_df = response_df[RESPONSE_THRESHOLD.values()] 
# response_df.columns = response_df.columns.astype(str)
response_df_org = response_df.copy()
response_df = response_df.div(response_df.sum(axis=1), axis=0) #normalize by total number of calls
response_df = response_df.merge(socio_df, how='inner', left_on='nbh_id', right_on='NBH_ID')#.drop(columns=['block_id'])
response_df.index = response_df['NBH_ID']

response_df
```

## Clustering

```python
'''
Clustering
'''
!rm -rf exports
!mkdir 'exports/'
!mkdir 'exports/KMeans'
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

import numpy as np
from tqdm.notebook import tqdm
import json
import matplotlib.pyplot as plt
distortions_call = []; distortions_dept = []; distortions_response = []; distortions_freq = []; distortions_socioeco = []; distortions_all = []
sil_call = []; sil_dept = []; sil_response = []; sil_freq = []; sil_socioeco = []; sil_all = []
sc = StandardScaler()

for i in tqdm(range(2,10)):
  '''
  Cluster by Call
  '''
  df = merged_df.loc[:,"Animals":"Water"]
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  # kmeans.labels_
  master_df['Cluster by Call Category'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Call Category').size().to_frame('count'))
  # distortions_call.append(kmeans.inertia_)
  # sil_call.append(silhouette_score(df, kmeans.labels_, metric = 'euclidean'))

  '''
  Cluster by Department
  '''
  df = merged_df.loc[:, 'City Services (dept)':'Water Services (dept)']
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  # kmeans.labels_
  master_df['Cluster by Department'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Category').size().to_frame('count'))
  distortions_dept.append(kmeans.inertia_)
  sil_dept.append(silhouette_score(df, kmeans.labels_, metric = 'euclidean'))

  '''
  Response Time (Days to Close) Clustering
  '''
  df = merged_df.iloc[:,:-5]
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  master_df['Cluster by Response Time'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Response Time').size().to_frame('count'))
  distortions_response.append(kmeans.inertia_)
  sil_response.append(silhouette_score(df, kmeans.labels_, metric = 'euclidean'))
  
  '''
  Call Frequency Clustering
  '''
  df = merged_df['Frequency'].to_frame('Frequency')
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  master_df['Cluster by Call Frequency'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Call Frequency').size().to_frame('count'))
  distortions_freq.append(kmeans.inertia_)
  sil_freq.append(silhouette_score(df , kmeans.labels_, metric = 'euclidean'))
  

  '''
  Socio Economic Clustering
  '''
  
  '''
    Population Attribute Group
  '''
  df = df_population.copy()
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(len(cluster_data))
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  # print(len(kmeans.labels_))
  master_df['Cluster by Population Attr Group'] = kmeans.labels_ + 1
  '''
    Economic Attribute Group
  '''
  df = df_economic.copy()
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  master_df['Cluster by Economic Attr Group'] = kmeans.labels_ + 1
  '''
    Ethnicity Attribute Group
  '''
  df = df_ethnicity.copy()
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  master_df['Cluster by Ethnicity Attr Group'] = kmeans.labels_ + 1



  df = master_df[['Cluster by Population Attr Group', 'Cluster by Economic Attr Group', 'Cluster by Ethnicity Attr Group']].astype(int)
  kmeans = KMeans(n_clusters=i, random_state=0).fit(df)
  kmeans.labels_
  master_df['Cluster by Socioeconomic Metrics'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Socioeconomic Metrics').size().to_frame('count'))

  distortions_socioeco.append(kmeans.inertia_)
  sil_socioeco.append(silhouette_score(df, kmeans.labels_, metric = 'euclidean'))

  '''
  Clustering by Crime
  '''
  df = merged_df['Crime Frequency'].to_frame('Frequency')
  X = np.nan_to_num(df.values)
  cluster_data = sc.fit_transform(X)
  # print(cluster_data[:5])
  kmeans = KMeans(n_clusters=i, random_state=0).fit(cluster_data)
  master_df['Cluster by Crime Frequency'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by Call Frequency').size().to_frame('count'))
  distortions_freq.append(kmeans.inertia_)
  sil_freq.append(silhouette_score(df , kmeans.labels_, metric = 'euclidean'))
  

  '''
  All-factor Clustering
  '''
  df = master_df[['Cluster by Call Category', 'Cluster by Department', 'Cluster by Response Time', 'Cluster by Call Frequency', 'Cluster by Socioeconomic Metrics', 'Cluster by Crime Frequency']]
  kmeans = KMeans(n_clusters=i, random_state=0).fit(df)
  master_df['Cluster by All Factors'] = kmeans.labels_ + 1
  # display(master_df.groupby('Cluster by All Factors').size().to_frame('count')).
  distortions_all.append(kmeans.inertia_)
  sil_all.append(silhouette_score(df, kmeans.labels_, metric = 'euclidean'))
  master_df = master_df.sort_values('Cluster by Call Category')

  master_df.to_csv('exports/KMeans/246NBH-{}clusters-KMeans.csv'.format(str(i)), index = False)
  with open('exports/KMeans/246NBH-{}clusters-KMeans.json'.format(i), 'w') as f:
    bg_dict = master_df.to_dict('index')
    bg_dict['Cluster_Total'] = i
    f.write(json.dumps(bg_dict))

  
  # break

```

```python
import pandas as pd

allinfo_df = pd.read_csv('/content/exports/246NBH-all_info.csv')
allinfo_df.index = allinfo_df['NBH_ID']
allinfo_df.head()
```

```python
import pandas as pd
import json
from tqdm.notebook import tqdm
cluster_columns = ['Cluster by Call Category', 'Cluster by Department', 
                   'Cluster by Response Time', 'Cluster by Call Frequency', 
                   'Cluster by Socioeconomic Metrics', 'Cluster by Crime Frequency', 'Cluster by All Factors']
for i in tqdm(range(2,10)):
  df = pd.read_csv('exports/KMeans/246NBH-{}clusters-KMeans.csv'.format(str(i)))

  max_num = df['Cluster by Call Category'].max()
  print(max_num)
  # display(df.describe())
  with open('exports/KMeans/246NBH-{}clusters-KMeans.json'.format(i), 'r') as f:
        clusters_ann = json.load(f)
        clusters_ann['Cluster_Profiles'] = {}
  for col in cluster_columns:
    print(col)
    columns = allinfo_df.columns
    cluster_profiles = []
    for i in range(1,max_num+1):
      # print(i)
      cluster_members = df[df[col] == i]['NBH_ID'].values.tolist()

      if 'Call Category' in col:
        sub_df = allinfo_df.copy().loc[:, 'Animals': 'Water']
      elif 'Department' in col:
        sub_df = allinfo_df.copy().loc[:, 'City Services (dept)': 'Water Services (dept)']
      elif 'Response Time' in col:
        sub_df = allinfo_df.copy().loc[:, 'On time (0-day delay)': '365+ -day window']
      elif 'Call Frequency' in col:
        sub_df = allinfo_df.copy().loc[:, 'Frequency'].to_frame('Frequency')
      elif 'Socioeconomic' in col:
        sub_df = allinfo_df.copy().loc[:, 'Total population': 'Doctorate degree']
      elif 'Crime' in col:
        sub_df = allinfo_df.copy().loc[:, 'Crime Frequency'].to_frame('Crime Frequency')
      else:
        sub_df = allinfo_df.copy().loc[:, 'Animals': 'Crime Frequency'].drop(['NBH_ID', 'NBH_NAME', 'EQUIVALENT_BLOCKGROUPS'], axis = 1)

      single_df = sub_df[sub_df.index.isin(cluster_members)]
      # display(single_df.head())
      if len(single_df) > 1:
        profile_df = single_df.describe(include='all').round(5)#.astype(int)
      else:
        profile_df = single_df.describe(include='all').fillna('N/A')

      if 'Call Category' in col or 'Department' in col:
        profile_df = profile_df.sort_values(by = 'mean', axis = 1, ascending = False)
      # display(profile_df)
      
      profile_dict = profile_df.to_dict()
      profile_dict['Cluster_ID'] = i
      cluster_profiles.append(profile_dict)
    clusters_ann['Cluster_Profiles'][col] = cluster_profiles
  with open('exports/KMeans/246NBH-{}clusters-KMeans.json'.format(i), 'w') as f:
        f.write(json.dumps(clusters_ann))
    # break
  # break
```

```python
metadata = {}
cols = allinfo_df.copy().loc[:, 'Animals': 'Water'].columns.tolist()
metadata['Categories'] = cols
cols = allinfo_df.copy().loc[:, 'City Services (dept)': 'Water Services (dept)'].columns.tolist()
metadata['Departments'] = cols
cols = allinfo_df.copy().loc[:, 'On time (0-day delay)': '365+ -day window'].columns.tolist()
metadata['Response Times'] = cols
cols = allinfo_df.copy().loc[:, 'Frequency'].to_frame('Frequency').columns.tolist()
metadata['Frequency'] = cols
cols = allinfo_df.copy().loc[:, 'Total population': 'Doctorate degree'].columns.tolist()
metadata['Socioeconomic Metrics'] = cols
cols = allinfo_df.copy().loc[:, 'Crime Frequency'].to_frame('Crime Frequency').columns.tolist()
metadata['Crime Frequency'] = cols
metadata

from glob import glob
import json
NBH_cluster_info = [metadata]
NBH_list = glob('./exports/KMeans/*.json')
for path in NBH_list:
  with open(path) as f:
    temp = json.load(f)
    keys = list(temp.keys())[:] + ['Cluster_Total']
    filtered_dict = {k: temp[k] for k in keys}
    NBH_cluster_info.append(filtered_dict)
  # break
# NBH_cluster_info = json.dumps(NBH_cluster_info)
```
